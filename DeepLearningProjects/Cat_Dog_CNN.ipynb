```python
import tensorflow as tf
import matplotlib.pyplot as plt
from tensorflow import keras
from keras import layers, models
from tensorflow import keras
import time

# Measure the start time for training
overall_start_time = time.time()

train_dir = '/content/drive/MyDrive/Cat-Dog/train'  # Adjust if the structure in the zip is different
test_dir = '/content/drive/MyDrive/Cat-Dog/test'

print(train_dir, test_dir)

# Constants
class_names = ['cat', 'dog']
num_epochs = 15
batch_size = 160

# Function to load datasets dynamically
def load_datasets(img_size):
    train_dataset = tf.keras.utils.image_dataset_from_directory(
        train_dir,
        image_size=(img_size, img_size),
        batch_size=batch_size,
        label_mode='int'
    ).map(lambda x, y: (x / 255.0, y))  # Normalize the pixel values

    test_dataset = tf.keras.utils.image_dataset_from_directory(
        test_dir,
        image_size=(img_size, img_size),
        batch_size=batch_size,
        label_mode='int'
    ).map(lambda x, y: (x / 255.0, y))  # Normalize the pixel values
    return train_dataset.prefetch(buffer_size=tf.data.AUTOTUNE), test_dataset.prefetch(buffer_size=tf.data.AUTOTUNE)

# Function to create models dynamically based on input configuration
def create_model(num_conv_layers, img_size):
    model = keras.models.Sequential()
    # Add convolutional and pooling layers
    for i in range(num_conv_layers):
        filters = 32 * (i + 1)
        model.add(keras.layers.Conv2D(
            filters=filters,
            kernel_size=(3, 3),
            strides=(1, 1),
            padding='same',
            activation='relu',
            input_shape=(img_size, img_size, 3) if i == 0 else None
        ))
        model.add(keras.layers.MaxPooling2D(pool_size=(2, 2), strides=(2, 2), padding='same'))

    # Add flatten and dense layers
    model.add(keras.layers.Flatten())
    model.add(keras.layers.Dense(64, activation='relu'))
    model.add(keras.layers.Dense(2, activation='softmax'))  # Binary classification

    # Compile the model
    model.compile(
        optimizer='adam',
        loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=False),
        metrics=['accuracy']
    )
    return model

# Test configurations
configurations = [
    {"num_conv_layers": 1, "img_size": 32},
    {"num_conv_layers": 1, "img_size": 64},
    {"num_conv_layers": 1, "img_size": 112},
    {"num_conv_layers": 2, "img_size": 32},
    {"num_conv_layers": 2, "img_size": 64},
    {"num_conv_layers": 2, "img_size": 112},
    {"num_conv_layers": 3, "img_size": 32},
    {"num_conv_layers": 3, "img_size": 64},
    {"num_conv_layers": 3, "img_size": 112}
]

# Run each configuration
for config in configurations:
    num_conv_layers = config["num_conv_layers"]
    img_size = config["img_size"]

    print(f"\nRunning configuration: {num_conv_layers} Conv Layers, Image Size {img_size}x{img_size}")

    # Load datasets with updated image size
    train_dataset, test_dataset = load_datasets(img_size)

    # Create and train the model
    model = create_model(num_conv_layers, img_size)
    model.summary()

    history = model.fit(
        train_dataset,
        epochs=num_epochs,
        validation_data=test_dataset,
        verbose=1
    )

    # Evaluate the model
    start_time = time.time()
    test_loss, test_acc = model.evaluate(test_dataset, verbose=2)
    evaluation_time = time.time() - start_time
    print(f"Evaluation completed in {evaluation_time:.2f} seconds.")
    print(f"Test accuracy: {test_acc:.2f}")

    # Plot training and validation accuracy
    plt.figure()
    plt.plot(history.history['accuracy'], label='Training Accuracy')
    plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
    plt.title(f"{num_conv_layers} Conv Layers, Image Size {img_size}x{img_size}")
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend(loc='lower right')
    plt.show()

# Calculate and print overall execution duration
overall_evaluation_time = time.time() - overall_start_time
print(f"Total execution time: {overall_evaluation_time:.2f} seconds.")
```
